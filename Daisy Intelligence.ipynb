```python
import matplotlib.pyplot as plt
def transform(track, handle):
    arr = []
    with open(track, "r") as f:
        next(f)
        for line in f:
            curv = float(line.strip())
            if curv != -1.0:
                vmax = (handle * curv/1000000)**0.5
                if vmax > 50:
                    vmax = 50
            else:
                vmax = 50
            arr.append(vmax)
    return arr
```

```python
track = transform('track_1.csv', 15)
#print(track)
#print(track)
for k in range(10):
    x = [i for i in range(k*100, (k+1)*100)]
    plt.figure(figsize=(20,20))
    plt.subplot(10, 1, k+1)
    plt.plot(x, track[k*100:(k+1)*100])
#     for a, b in zip(x, track[k*100:(k+1)*100]):
#         plt.text(a, b, str(b))
    #print(len(track[k*10:(k+1)*10]))

# plt.show()
# plt.figure(figsize=(20,10))
# plt.plot([i for i in range(1000)], track)
```
